# Plotting distribution of correlated random variables (Problem 10.2b)



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

x1 = np.linspace(-3, 3, 100)
x2 = np.linspace(-3, 3, 100)

def gauss(epsilon):
    X1, X2 = np.meshgrid(x1, x2)
   
    arg = -1/(2 * (1 - epsilon**2)) * (X1**2 - 2*epsilon*X1*X2 + X2**2)
    return 1 / (2*np.pi*np.sqrt(1 - epsilon**2)) * np.exp(arg)
   
```

```python
epsilon = -0.5
plt.imshow(gauss(epsilon), extent=[-3, 3, -3, 3])
```
